## Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a powerful data visualization library for Python that is used to create static, animated, and interactive visualizations in Python. It provides a wide variety of functions and tools for creating different types of plots and charts, ranging from simple line and scatter plots to complex 3D visualizations.

Matplotlib is used for data analysis, data visualization, scientific plotting, and more. It allows users to easily create visual representations of their data, which can help to identify patterns, trends, and relationships within the data. Matplotlib can be used in a wide variety of fields, including science, engineering, finance, and economics.

Five types of plots that can be created using the Pyplot module of Matplotlib are:

Line plots: used to display data as a series of points connected by straight lines.
Scatter plots: used to display the relationship between two variables as a collection of points.
Bar plots: used to display data as bars of different heights or lengths.
Histograms: used to display the distribution of a dataset by dividing it into a set of intervals and counting the number of data points that fall into each interval.
Pie charts: used to display data as a circle divided into sections, where each section represents a proportion of the whole.

## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

A scatter plot is a type of plot that displays the relationship between two variables by plotting their values as a collection of points. Each point represents a single observation or data point, with the x-coordinate representing the value of one variable and the y-coordinate representing the value of the other variable.

Scatter plots are used to visualize the relationship between two variables and to identify patterns or trends in the data. They can be used to identify correlations between variables, outliers, clusters or groups of data points, and other patterns in the data.

```python
np.random.seed(3)
x = 3+np.random.normal(0,2,50)
y = 3+np.random.normal(0,2,len(x))
plt.scatter(x,y)
plt.title("This is a scatter plot")
plt.xlabel("This is X axis")
plt.ylabel("this is Y axis")
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

import numpy as np
1. For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

2. For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

3. For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

4. For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for the four lines
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create the subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# Plot line 1 in the top left subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# Plot line 2 in the top right subplot
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# Plot line 3 in the bottom left subplot
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Plot line 4 in the bottom right subplot
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Add a title for the entire figure
fig.suptitle('Four line plots using subplot()')

# Display the plot
plt.show()
```

A bar plot, also known as a bar chart or bar graph, is a graphical representation of data in which rectangular bars of equal width are used to show the values of a categorical variable. The height or length of each bar corresponds to the value or frequency of the variable being plotted.

Bar plots are useful when we want to compare the values of different categories or groups. They are often used to display data that is nominal or ordinal in nature. For example, we might use a bar plot to compare the number of students in different grades, or the frequency of different types of vehicles on a road.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
```

```python
plt.barh(company,profit)
```
